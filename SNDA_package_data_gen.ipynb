```python
def get_winning_team(row):
  if row['home_score'] > row['away_score']:
    return row['home_team']
  elif row['home_score'] < row['away_score']:
    return row['away_team']
  else:
    return 'Draw'

df['winning_team'] = df.apply(get_winning_team, axis=1)
```

```python
def get_losing_team(row):
  if row['home_score'] < row['away_score']:
    return row['home_team']
  elif row['home_score'] > row['away_score']:
    return row['away_team']
  else:
    return 'Draw'

df['losing_team'] = df.apply(get_losing_team, axis=1)
```

```python
df.head()
```

```python
outliers=df[df['winning_team']=='Draw']
outliers
```

```python
df.drop(outliers.index,inplace=True,axis=0)
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
teams=list(df['home_team'].unique())
teams+=list(df['away_team'].unique())
teams=list(set(teams))
```

```python
teams
```

```python
len(teams)
```

```python
weights={}
for i in teams:
  weights[i]={}
  for j in teams:
    if i!=j:
      weights[i][j]=0
weights[list(weights.keys())[0]]
```

```python
for i in teams:
  temp=df.query('losing_team==@i')
  total_meets=len(temp)
  for j in teams:
    if i!=j:
      if total_meets==0:
          weights[i][j]=0
      else:
        weights[i][j]=len(temp.query('home_team==@j or away_team==@j'))/total_meets
weights[list(weights.keys())[0]]
```

```python
sum(weights[list(weights.keys())[0]].values())
```

Weight denotes the ratio of a node i losing with node j to the total matches they have lost

```python
final_network_data=pd.DataFrame(columns=['source','target','weight'])
set_final=set()
for i in weights.keys():
  for j in weights[i].keys():
    if weights[i][j]!=0:
      set_final.add((i,j,weights[i][j]))
set_final
```

```python
  for i, j, k in set_final:
  final_network_data = pd.concat([final_network_data, pd.DataFrame({'source': [i], 'target': [j], 'weight': [k]})], ignore_index=True)
final_network_data
```
